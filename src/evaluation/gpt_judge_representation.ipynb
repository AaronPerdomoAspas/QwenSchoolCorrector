```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# === Ruta al JSON que contiene todas las muestras
with open("F:/Usuario/Desktop/Final/g_eval_resultados_unificado.json", "r", encoding="utf-8") as f:
    data = json.load(f)
df = pd.DataFrame(data)
```

```python
metrics = ["coherence", "consistency", "fluency", "relevance"]

# === Calcular medias
promedios_base = df[[f"{m}_base" for m in metrics]].mean()
promedios_fine = df[[f"{m}_finetuned" for m in metrics]].mean()
```

```python
# === Crear DataFrame comparativo
df_comp = pd.DataFrame({
    "Modelo base": promedios_base.values,
    "Modelo fine-tuneado": promedios_fine.values
}, index=metrics)
```

```python
# === Visualizar
df_comp.plot.bar(figsize=(8, 5), rot=0)
plt.title("Comparación de métricas G-Eval entre modelos")
plt.ylabel("Puntuación media")
plt.ylim(0, 5.5)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

# Mostrar resultados redondeados
df_comp.round(2)
```
